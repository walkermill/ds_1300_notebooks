```python
import pandas as pd
data = pd.read_csv('nba_19_20_pg-Copy1.csv')
data[['Pos', 'Player', 'PTS']].head(100)
#data
```

```python
import matplotlib.pyplot as plt
```

```python
group = data.groupby(['Pos']).sum()
```

```python
players = group.loc[['C', 'PF', 'PG', 'SF', 'SG']]
players
```

```python
sum(players['3P'])
#print(players['3P'][0])
```

```python
labels = 'C', 'PF', 'PG', 'SF', 'SG'
sizes = [(players['3P'][0])/sum(players['3P']), (players['3P'][1])/sum(players['3P']), (players['3P'][2])/sum(players['3P']), (players['3P'][3])/sum(players['3P']), (players['3P'][4])/sum(players['3P'])]



fig, ax = plt.subplots()
ax.pie(sizes, labels=None, autopct='%1.1f%%')
plt.title('% of 3 Points Made by Position (2019-2020)')
ax.legend(loc=1, labels=labels)
plt.savefig('%_3_Points_Made_by_Position).pdf')
plt.show()
```

```python
# My plot shows the percent of 3 pointers that each position made in the 2019-2020 basketball 
# season out of the total 3 pointers made.
```

```python
# I chose to make the plot the way I did because I thought it was the easiest way to visualize the data.
# You can easily see the percentage that each position made out of the total 3 pointers, what positions correlate
# to which percentages, and what that percent visually looks like compared to the other positions. I also like that
# it is ball shaped, so you could maybe swap out the pie chart graphic for a basketball and explode each slice.
```
